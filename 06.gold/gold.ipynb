### Полнота обогащения

Теперь проверим имеющуюся у нас формулу полноты обогащения. Напишем функцию для вычисления полноты и сразу проверим её на обучающей выборке.

```python
# функция принимает на вход 3 pd.Series

def recovery_calc (concentrate, feed, tail):
    
    recovery = (concentrate*(feed-tail))/(feed*(concentrate-tail))   
    
    return recovery*100
```

```python
rougher_calculated_recovery = recovery_calc(df_train['rougher.output.concentrate_au'],\
                                            df_train['rougher.input.feed_au'],\
                                            df_train['rougher.output.tail_au'])

mae(rougher_calculated_recovery,df_train['rougher.output.recovery'])
```

#### Вывод:

Среднее абсолютное отклонение = ~10^-14, крайне мало, учитывая, что Recovery принимает значения от 1 до 100. Формула верна, можно ей пользоваться в дальнейших расчётах.

### Динамика концентраций металлов на различных этапах очистки.

Изменение концентрации металлов на разных этапах проще всего оценить с помощью визуализации.  

Для этого напишем функцию для элемента, которая будет строить график с четырьмя сериями данных - по каждой стадии процесса.

Мы уже ранее выделяли стадии. Отбросим из них хвостовые.

```python
for stage in stages_list:
    if '.tail' in stage:
        stages_list.remove(stage)
# почему-то по итогам одной итерации не удалось удалить одну из хвостовых стадий, задублировал    
for stage in stages_list:    
    if '.tail' in stage:
        stages_list.remove(stage)

stages_list      
```

```python
# Отсортируем стадии

stages_list.sort(reverse = True)
display(stages_list)

# Поменяем два элемента индексами

a = stages_list.index('rougher.input.feed_')
b = stages_list.index('rougher.output.concentrate_')

stages_list[a],stages_list[b] = stages_list[b], stages_list[a]

display(stages_list)
```

```python
def element_plot(element):
    
    plt.style.use('seaborn-paper')
    plt.figure(figsize = (12,6))
    plt.title(f"Постадийная динамика концентрации {element.upper()}")
              
    for stage in stages_list:
        column = (stage+element)
        plt.hist(df_full[column], alpha=0.5, label = column, bins = 200)
    
    plt.grid()    
    plt.legend(loc='upper right')    
    plt.show()
```

```python
for element in ['au','ag','pb']:
    element_plot(element)
```

#### Вывод:

**- Золото:**
  - Концентрация золота возрастает с каждым этапом обогащения. 

**- Серебро:**
  - Концентрация серебра увеличивается только после флотации, далее - падает с каждым этапом обогащения.
 
**- Свинец:**
  - Концентрация свинца возрастает по итогам флотации и незначительно колеблется на последующих этапах обогащения.

### Подготовка скорера и выборок

Для начала напишем функцию для подчета целевой метрики SMAPE и создадим на основе этой функции скорер для подбора модели.

```python
def SMAPE_weighted(target, predictions):
    
    target = target.to_numpy()
    
    smape = (1/len(target)) * sum(abs(target - predictions)/((abs(target)+abs(predictions))/2))
    
    score = 25*smape[0]+75*smape[1]
                                      
    return score
```

```python
smape_scorer = make_scorer(SMAPE_weighted, greater_is_better=False)
```

Теперь отбросим ненужные столбцы, разобьем данные на выборки, стандартизуем признаки и приступим к подбору модели.
